```python
import datetime
now = datetime.datetime.now()
df_train['Year'] = df_train['Year'].apply(lambda x : now.year - x)
df_test['Year'] = df_test['Year'].apply(lambda x : now.year - x)
```

```python
df_train=df_train.drop(columns='Location',axis=1)
df_test=df_test.drop(columns='Location',axis=1)
```

```python
df_train = pd.get_dummies(df_train, drop_first=True)
df_test = pd.get_dummies(df_test, drop_first=True)
```

```python
df_train
```

```python

X = df_train[['Year', 'Kilometers_Driven', 'Mileage', 'Engine', 'Power', 'Seats',
       'New_Price', 'Name_BMW', 'Name_Bentley', 'Name_Chevrolet',
       'Name_Datsun', 'Name_Fiat', 'Name_Ford', 'Name_Honda',
       'Name_Hyundai', 'Name_ISUZU', 'Name_Isuzu', 'Name_Jaguar', 'Name_Jeep',
       'Name_Land', 'Name_Mahindra', 'Name_Maruti', 'Name_Mercedes-Benz',
       'Name_Mini', 'Name_Mitsubishi', 'Name_Nissan', 
       'Name_Porsche', 'Name_Renault', 'Name_Skoda', 'Name_Tata',
       'Name_Toyota', 'Name_Volkswagen', 'Name_Volvo', 'Fuel_Type_Diesel',
       'Fuel_Type_LPG', 'Fuel_Type_Petrol', 'Transmission_Manual',
       'Owner_Type_Fourth & Above', 'Owner_Type_Second', 'Owner_Type_Third']]

y = df_train['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

```python
xgb=xgboost.XGBRegressor(colsample_bytree=0.8, subsample=0.5,
                             learning_rate=0.05, max_depth=3, 
                             min_child_weight=1.8, n_estimators=2000,
                             reg_alpha=0.1, reg_lambda=0.3, gamma=0.01, 
                             silent=1, random_state =7, nthread = -1)
```

```python
xgb.fit(X_train,y_train)
```

```python
y_pred=xgb.predict(X_test)
print(y_pred)
```

```python
df_test_x=df_test[['Year', 'Kilometers_Driven', 'Mileage', 'Engine', 'Power', 'Seats',
       'New_Price', 'Name_BMW', 'Name_Bentley', 'Name_Chevrolet',
       'Name_Datsun', 'Name_Fiat', 'Name_Ford', 'Name_Honda',
       'Name_Hyundai', 'Name_ISUZU', 'Name_Isuzu', 'Name_Jaguar', 'Name_Jeep',
       'Name_Land', 'Name_Mahindra', 'Name_Maruti', 'Name_Mercedes-Benz',
       'Name_Mini', 'Name_Mitsubishi', 'Name_Nissan', 
       'Name_Porsche', 'Name_Renault', 'Name_Skoda', 'Name_Tata',
       'Name_Toyota', 'Name_Volkswagen', 'Name_Volvo', 'Fuel_Type_Diesel',
       'Fuel_Type_LPG', 'Fuel_Type_Petrol', 'Transmission_Manual',
       'Owner_Type_Fourth & Above', 'Owner_Type_Second', 'Owner_Type_Third']]
```
